```python
import pandas as pd
```

```python
df = pd.read_csv('titanic.csv')
df.head()
```

```python
df.isnull().sum()
```

```python
df["Age"] = df["Age"].fillna(df["Age"].median())
df["Fare"] = df["Fare"].fillna(df["Fare"].median())
df["Cabin"] = df["Cabin"].fillna("Unknown")
```

```python
df['PassengerClass'] = df['Pclass'].map({1: "First", 2: "Second", 3: "Third"})
df['Survived_Label'] = df['Survived'].map({0: "Dead", 1: "Alive"})
df["Sex_Label"] = df["Sex"].str.capitalize()
df["EmbarkedPort"] = df["Embarked"].map({"C": "Cherbourg", "Q": "Queenstown", "S": "Southampton"})
```

```python
df.to_csv('titanic_cleaned.csv', index = False)
```
